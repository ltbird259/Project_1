```python
#Import 
import requests
from config import api_key
from pprint import pprint
from formatjh import *
import pandas as pd
url = f"http://www.omdbapi.com/?apikey={api_key}&t="
```

```python
#Import CSV files and make the lists
horror = pd.read_csv("Horror.csv")
comedy = pd.read_csv("Comedy.csv")
horror_list= horror["Title"]
comedy_list = comedy["Title"]
horror_list_sample = ["Aliens","Despicable Me","I Spit on Your Grave"]
```

```python
#Get information that we need
#create lists that will become our movie's data
#This block of code will be for our horror movie data
movie_response = []
imdb_id = []
title = []
media_type = []
year = []
country = []
genre = []
awards = []
metascore = []
production = []
rated = []
imdb_rating = []
imdb_votes = []
for movie in horror_list:
    autoformat(f"Making request for movie: {movie}")
    
    #save post's JSON
    try:
        result = requests.get(url + str(movie)).json()
        movie_response.append(result)
        #assign values to each list as we interate through making the calls
        imdb_id.append(result['imdbID'])
        title.append(result['Title'])
        media_type.append(result['Type'])
        year.append(result['Year'])
        country.append(result['Country'])
        genre.append(result['Genre'])
        awards.append(result['Awards'])
        metascore.append(result['Metascore'])
        production.append(result['Production'])
        rated.append(result['Rated'])
        imdb_rating.append(result['imdbRating'])
        imdb_votes.append(result['imdbVotes'])
        
    except KeyError:
         pass
         autoformat(f"{movie} failed to load")
        
autoformat("Download Complete!!")
```

```python
horror_dict = {
    "IMDB ID" : imdb_id,
    "Title" : title,
    "Type of Media" : media_type,
    "Year" : year,
    "Country":country,
    "Genre":genre,
    "Awards":awards,
    "Production Company":production,
    "Rating":rated,
    "Metascore":metascore,
    "IMDB Rating":imdb_rating,
    "# of Votes":imdb_votes    
}
```
