```python
#Import 
import requests
from config import api_key
from pprint import pprint
from formatjh import *
import pandas as pd
url = f"http://www.omdbapi.com/?apikey={api_key}&t="
```

```python
#Import CSV files and make the lists
horror = pd.read_csv("Horror.csv")
horror_list= horror["Title"]
horror_list_sample = ["Aliens","Despicable Me","I Spit on Your Grave"]
```

```python
#Get information that we need
#create lists that will become our movie's data
#This block of code will be for our horror movie data
movie_response = []
for movie in horror_list:
    autoformat(f"Making request for movie: {movie}")
    
    #save post's JSON
    try:
        result = requests.get(url + str(movie)).json()
        movie_response.append(result)
    except KeyError:
        pass
        autoformat(f"{movie} failed to load")
        
autoformat("Download Complete!!")
```

```python
movie_response
```

```python
#assign values to each list as we interate through making the calls
imdb_id = []
title = []
media_type = []
year = []
country = []
genre = []
awards = []
metascore = []
production = []
rated = []
imdb_rating = []
imdb_votes = []

for x in movie_response:
    try:
        imdb_id.append(x['imdbID'])
    except KeyError:
        imdb_id.append("NA")
    try:
        title.append(x['Title'])
    except KeyError:
        title.append("NA")
    try:
        media_type.append(x['Type'])
    except KeyError:
        media_type.append("NA")
    try:
        year.append(x['Year'])
    except KeyError:
        year.append("NA")
    try:
        country.append(x['Country'])
    except KeyError:
        country.append("NA")
    try:
        genre.append(x['Genre'])
    except KeyError:
        genre.append("NA")
    try:    
        awards.append(x['Awards'])
    except KeyError:
        awards.append("NA")
    try:
        metascore.append(x['Metascore'])
    except KeyError:
        metascore.append("NA")
    try:
        production.append(x['Production'])
    except KeyError:
        production.append("NA")
    try:
        rated.append(x['Rated'])
    except KeyError:
        rated.append("NA")
    try:
        imdb_rating.append(x['imdbRating'])
    except KeyError:
        imdb_rating.append("NA")
    try:
        imdb_votes.append(x['imdbVotes'])
    except KeyError:
        imdb_votes.append("NA")
        
```

```python
horror_dict = {
    "IMDB ID" : imdb_id,
    "Title" : title,
    "Type of Media" : media_type,
    "Year" : year,
    "Country":country,
    "Genre":genre,
    "Awards":awards,
    "Production Company":production,
    "Rating":rated,
    "Metascore":metascore,
    "IMDB Rating":imdb_rating,
    "# of Votes":imdb_votes    
}
```

```python
horror_df = pd.DataFrame(horror_dict)
horror_df.head()
```

```python
horror_df.to_csv("Horror_dataframe.csv", index=False, header=True)
```
